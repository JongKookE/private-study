## import library 


```python
import numpy as np
import matplotlib.pyplot as plt
```

### show graph

```python
x = np.linspace(0, 2*np.pi, 100)
```

```python
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
plt.plot(x,y1, label = "sin")
plt.plot(x,y2, label = "cos")
plt.legend()
plt.show()
```
